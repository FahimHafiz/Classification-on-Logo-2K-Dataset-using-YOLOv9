```python
import os

from ultralytics import YOLO
# Load a model
model = YOLO("/content/drive/MyDrive/yolov9c-cls.yaml")  # build a new model for logo classification using custom YAML

```

```python
# Use the model
results = model.train(data=DATA_DIR, epochs=50, imgsz=256)  # train the model
```

```python
# Validation
best_model = YOLO('/content/runs/classify/train/weights/best.pt')


# Testing the best Model
test_results = best_model.predict(source='/content/drive/MyDrive/Logo-2k+-subclass/train_and_test/test/CTC/13.jpg', stream=False, conf=0.25, save=True)

class_names = test_results[0].names
class_probabilities_predict = test_results[0].probs # Probabilities associated with all the classes predicted by our model
predicted_class = class_names[np.argmax(class_probabilities_predict.data.cpu()).item()]

print(f"Predicted class for the test image from CTC logo: {predicted_class}")
```
